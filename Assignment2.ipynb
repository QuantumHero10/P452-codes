```python
import numpy as np
import mylibrary.cjg_fly as cjg_fly # Conjugate Gradient inverter method
import mylibrary.mm as mm # Importing matrix multiplication function

def delta(x,y):
    if x == y:
        return 1
    else:
        return 0
def A(x,y): # Matrix calculated using function (not stored)
    mu=1
    m=0.2
    return 0.5*(delta(x+mu,y)+delta(x-mu,y)-(2*delta(x,y))) + ((m**2)*delta(x,y))

n=50
A1=np.zeros((n,n))
for i in range(0,n):
    for j in range(0,n):
        A1[i][j]=A(i,j)

A_inverse,R = cjg_fly.matrix_inverter(A1) # Finding the inverse matrix

print("\nThe inverse matrix(for tolerance 10^-6) is :\n")
cjg_fly.printarray(A_inverse)

print("\nMultipling A matrix and A_inverse matrix we get :\n")  # Checking the correctness of inverse
M = mm.main(np.matrix(A1),np.matrix(A_inverse))
cjg_fly.printarray(np.round(M,decimals=4)) # Rounding off the elements to 4 decimal places

# By rounding off to 4 decimal places we get the multiplication result equal to identity.
# Which shows that the inverse matrix is accurate.
# Open the ouput in scrollable element mode to view full output.

import matplotlib.pyplot as plt
X=np.zeros((len(R)))
for i in range (0,len(R)):
    X[i] = i
# Plot of Iteration vs Residue
plt.plot(X,R)
plt.xlabel("Iteration step")
plt.ylabel("Residue")
plt.title("Residue Convergence")
plt.show()

# The plot is at bottom of the page.



```
