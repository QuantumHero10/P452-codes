```python
import numpy as np
import mylibrary.cjg_fly as cjg_fly
import mylibrary.congradinv as cgi
import mylibrary.mm as mm # Importing matrix multiplication function

def delta(x,y):
    if x == y:
        return 1
    else:
        return 0
def A(x,y):
    if x == 0 or x == (2*y):
        mu = int(y)
    elif y == 0 or y == (2*x):
        mu = int(x)
    else:
        mu = int(x)
    m=0.2
    return 0.5*(delta(x+mu,y)+delta(x-mu,y)-(2*delta(x,y))) + ((m**2)*delta(x,y))

n=50
A1=np.zeros((n,n))
for i in range(0,n):
    for j in range(0,n):
        A1[i][j]=A(i,j)

cjg_fly.printarray(A1)
A_inverse,R = cjg_fly.matrix_inverter(A1) # Finding the inverse matrix

cjg_fly.printarray(A_inverse)

import matplotlib.pyplot as plt

# Plot heatmap
plt.imshow(R, cmap='viridis', interpolation='nearest')
plt.colorbar()  # Add color bar
plt.show()

#print("\nMultipling A matrix and A_inverse matrix we get :\n")
#M = mm.main(np.matrix(A1),np.matrix(A_inverse))
#cjg_fly.printarray(np.round(M,decimals=4)) # Rounding off the elements to 4 decimal places
```
